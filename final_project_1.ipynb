```python
#now lets check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#for best analysis i need to drop  columns with more than 20% missing values
df.dropna(axis=1, thresh=int(0.8*len(df)), inplace=True)  
#drop rows with fully no values
df.dropna(axis=0, how='all', inplace=True)  
```

```python
#replacing missing value in column with o
df.fillna(value=0, inplace=True)
df.isnull().sum()
```

```python
#recheck for the shape of dataset
df.shape
```

```python
# Find the number of unique values in each column
unique_counts = df.nunique()

# Compare unique counts with the number of rows
unique_columns = unique_counts[unique_counts == len(df)]

# Display columns with unique values
print("Columns with unique values:")
print(unique_columns)
```

```python
#check for inconsistency of datatypes
df.apply(type).value_counts()  # Show the count of each type in a column
```

```python
#checking for values in event column_dates
df['Event.Date']
```

```python
#Convert 'Event.Date' to datetime
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')

#Filter events after the year 2000
df_after_1990 = df[df['Event.Date'].dt.year > 2000]

#  let me drop rows with invalid dates (NaT)
df_after_1990_clean = df_after_1990.dropna(subset=['Event.Date'])

 #Handle other missing values

df_after_1990_clean['Injury.Severity'] = pd.to_numeric(df_after_1990_clean['Injury.Severity'], errors='coerce')
df_after_1990_clean['Injury.Severity'].fillna(df_after_1990_clean['Injury.Severity'].mean(), inplace=True)

# Verify the cleaned DataFrame
print(df_after_1990_clean.shape)
print(df_after_1990_clean.head())

```

```python
#drop country not equal to USA
strictly_USA = df_after_1990_clean.drop(index=df_after_1990_clean[df_after_1990_clean['Country'] != 'United States'].index).reset_index(drop=True)

del df_after_1990_clean['Country']
strickly_USA
```

```python
#changing data type for column 
strictly_USA['Number.of.Engines'] = strictly_USA['Number.of.Engines'].astype('float64')
print('successfully changed')
```

Data Visualization

```python
#Event.Date Distribution (Line plot by year)
plt.figure(figsize=(12, 6))
df['Event.Date'].dt.year.value_counts().sort_index().plot(kind='line', color='b')
plt.title('Number of Events Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Events')
plt.grid(True)
plt.show()
```

```python
#Injury Severity (Bar plot)
plt.figure(figsize=(12, 4))
sns.countplot(x='Injury.Severity', data=df, palette='Set2')
plt.title('Injury Severity Count')
plt.xlabel('Injury Severity')
plt.ylabel('Count of Events')
plt.show()
```

```python
#Box Plot of Total Injuries (Fatal, Serious, Minor)
injury_columns = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries']
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[injury_columns], palette='Set3')
plt.title('Distribution of Injury Counts')
plt.ylabel('Number of Injuries')
plt.show()
```

```python

#Histogram of Total Uninjured
plt.figure(figsize=(10, 6))

df['Total.Uninjured'].plot(kind='hist', bins=50, color='b', alpha=0.7)
plt.title('Distribution of Total Uninjured')
plt.xlabel('Total Uninjured')
plt.ylabel('Frequency')
plt.show()
```


Conclusion:
In conclusion, the visualizations suggest that while most aviation events involve low to no injuries, there are notable outliers with serious and fatal injuries. Trends over time can provide insights into potential safety improvements or deteriorations. By focusing on the outliers and years with high injury counts, targeted safety measures can be developed to mitigate risks and reduce fatalities and serious injuries in future incidents.
